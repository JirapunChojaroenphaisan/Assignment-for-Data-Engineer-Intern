```python
import pandas as pd
import requests

url = "https://randomuser.me/api"

df = pd.DataFrame()

for i in range(20):
    response = requests.get(url)
    data = response.json()
    df = df.append(pd.json_normalize(data['results']), ignore_index=True)





```

```python
df = df[['gender', 'name.first', 'name.last']]
df
```

```python

df_gender = pd.DataFrame()

for index, row in df.iterrows():
    name = row['name.first']
    url = f"https://api.genderize.io/?name={name}"
    response = requests.get(url)
    data = response.json()
    df_gender = df_gender.append(pd.json_normalize(data), ignore_index=True)

df_gender
```

```python
df_final = df.merge(df_gender, left_on='name.first', right_on='name')
```

```python

df_final[['gender_x','name.first','name.last','gender_y','probability']]
```

```python
df_final['same_gender'] = df_final['gender_x'] == df_final['gender_y']
```

```python
df_final = df_final.rename(columns={'name.first': 'first_name', 'name.last': 'last_name', 'gender_x': 'gender(actual)','gender_y': 'gender(predict)'})
```

```python
df_final = df_final[['first_name','last_name','gender(actual)','gender(predict)','probability','same_gender']].style.hide_index()
```
